```python
df = pd.read_csv("data.csv")
df2 = pd.read_csv('raw_total_fight_data.csv', delimiter=';')

df2['Winner By Corner'] = df['Winner']
df2.rename(columns={'Fight_type': 'Weight_class'}, inplace=True)
# df2['weight_class'] = df2['Fight_type']
# df2 = df2.drop(columns=['Fight_type'])
df2['title_bout'] = df['title_bout']
df2['B_age'] = df['B_age']
df2['R_age'] = df['R_age']
df2['B_Stance'] = df['B_Stance']
df2['R_Stance'] = df['R_Stance']
df2['Referee'].fillna('Unknown', inplace=True)
df2['Winner'].fillna('Draw', inplace=True)
df2['B_Height_cms'] = df['B_Height_cms']
df2['R_Height_cms'] = df['R_Height_cms']
df2['B_current_win_streak'] = df['B_current_win_streak']
df2['R_current_win_streak'] = df['R_current_win_streak']
age = df2['B_age'].isnull() | df2['R_age'].isnull()
df2.loc[age, ['B_age', 'R_age']] = 28.0
df2['B_age'] = df2['B_age'].astype(int)
df2['R_age'] = df2['R_age'].astype(int)
stance = df2['B_Stance'].isnull() | df2['R_Stance'].isnull()
df2.loc[stance, ['B_Stance', 'R_Stance']] = 'Unknown'
df2['R_Height_cms'].fillna(178, inplace=True)
df2['B_Height_cms'].fillna(178, inplace=True)
df2['R_SIG_STR.'] = df2['R_SIG_STR.'].str.extract(r'(\d+) of').astype(int)
df2['B_SIG_STR.'] = df2['B_SIG_STR.'].str.extract(r'(\d+) of').astype(int)
df2['B_TOTAL_STR.'] = df2['B_TOTAL_STR.'].str.extract(r'of (\d+)').astype(int)
df2['R_TOTAL_STR.'] = df2['R_TOTAL_STR.'].str.extract(r'of (\d+)').astype(int)
df2.rename(columns={'R_TD': 'R_SIG_TD'}, inplace=True)
df2.rename(columns={'B_TD': 'B_SIG_TD'}, inplace=True)
df2['R_TOTAL_TD'] = df2['R_SIG_TD'].str.extract(r'of (\d+)').astype(int)
df2['B_TOTAL_TD'] = df2['B_SIG_TD'].str.extract(r'of (\d+)').astype(int)
df2['B_SIG_TD'] = df2['B_SIG_TD'].str.extract(r'(\d+) of').astype(int)
df2['R_SIG_TD'] = df2['R_SIG_TD'].str.extract(r'(\d+) of').astype(int)
df2['Format'] = df2['Format'].str.replace(r'\(\d-\d-\d-\d-\d\)', '', regex=True)
df2['Format'] = df2['Format'].str.replace(r'\(\d-\d-\d\)', '', regex=True)


df2['R_ATT_HEAD'] = df2['R_HEAD'].str.extract(r'of (\d+)').astype(int)
df2['B_ATT_HEAD'] = df2['B_HEAD'].str.extract(r'of (\d+)').astype(int)
df2.rename(columns={'R_HEAD': 'R_LANDED_HEAD'}, inplace=True)
df2.rename(columns={'B_HEAD': 'B_LANDED_HEAD'}, inplace=True)
df2['B_LANDED_HEAD'] = df2['B_LANDED_HEAD'].str.extract(r'(\d+) of').astype(int)
df2['R_LANDED_HEAD'] = df2['R_LANDED_HEAD'].str.extract(r'(\d+) of').astype(int)

df2['R_ATT_BODY'] = df2['R_BODY'].str.extract(r'of (\d+)').astype(int)
df2['B_ATT_BODY'] = df2['B_BODY'].str.extract(r'of (\d+)').astype(int)
df2.rename(columns={'R_BODY': 'R_LANDED_BODY'}, inplace=True)
df2.rename(columns={'B_BODY': 'B_LANDED_BODY'}, inplace=True)
df2['B_LANDED_BODY'] = df2['B_LANDED_BODY'].str.extract(r'(\d+) of').astype(int)
df2['R_LANDED_BODY'] = df2['R_LANDED_BODY'].str.extract(r'(\d+) of').astype(int)

df2['R_ATT_LEG'] = df2['R_LEG'].str.extract(r'of (\d+)').astype(int)
df2['B_ATT_LEG'] = df2['B_LEG'].str.extract(r'of (\d+)').astype(int)
df2.rename(columns={'R_LEG': 'R_LANDED_LEG'}, inplace=True)
df2.rename(columns={'B_LEG': 'B_LANDED_LEG'}, inplace=True)
df2['B_LANDED_LEG'] = df2['B_LANDED_LEG'].str.extract(r'(\d+) of').astype(int)
df2['R_LANDED_LEG'] = df2['R_LANDED_LEG'].str.extract(r'(\d+) of').astype(int)

df2['R_ATT_CLINCH'] = df2['R_CLINCH'].str.extract(r'of (\d+)').astype(int)
df2['B_ATT_CLINCH'] = df2['B_CLINCH'].str.extract(r'of (\d+)').astype(int)
df2.rename(columns={'R_CLINCH': 'R_LANDED_CLINCH'}, inplace=True)
df2.rename(columns={'B_CLINCH': 'B_LANDED_CLINCH'}, inplace=True)
df2['B_LANDED_CLINCH'] = df2['B_LANDED_CLINCH'].str.extract(r'(\d+) of').astype(int)
df2['R_LANDED_CLINCH'] = df2['R_LANDED_CLINCH'].str.extract(r'(\d+) of').astype(int)

df2['R_ATT_GROUND'] = df2['R_GROUND'].str.extract(r'of (\d+)').astype(int)
df2['B_ATT_GROUND'] = df2['B_GROUND'].str.extract(r'of (\d+)').astype(int)
df2.rename(columns={'R_GROUND': 'R_LANDED_GROUND'}, inplace=True)
df2.rename(columns={'B_GROUND': 'B_LANDED_GROUND'}, inplace=True)
df2['B_LANDED_GROUND'] = df2['B_LANDED_GROUND'].str.extract(r'(\d+) of').astype(int)
df2['R_LANDED_GROUND'] = df2['R_LANDED_GROUND'].str.extract(r'(\d+) of').astype(int)

df2 = df2.drop(['R_DISTANCE', 'B_DISTANCE'], axis=1)

df2['date'] = pd.to_datetime(df2['date'])
df2['Day'] = df2['date'].dt.day
df2['Month'] = df2['date'].dt.month_name()
df2['Year'] = df2['date'].dt.year

df2['R_SIG_STR_pct'] = df2['R_SIG_STR_pct'].str.replace('%', '')
df2['B_SIG_STR_pct'] = df2['B_SIG_STR_pct'].str.replace('%', '')
df2['R_TD_pct'] = df2['R_TD_pct'].str.replace('%', '')
df2['B_TD_pct'] = df2['B_TD_pct'].str.replace('%', '')
df2['R_TD_pct'] = df2['R_TD_pct'].replace('---', 0)
df2['B_TD_pct'] = df2['B_TD_pct'].replace('---', 0)

df2['R_CTRL'] = df2['R_CTRL'].replace('--', "0:00")
df2['B_CTRL'] = df2['B_CTRL'].replace('--', "0:00")
```

```python
import pandas as pd
import plotly.express as px

# Assuming you have imported the necessary libraries and loaded the 'fights' DataFrame

# Create a copy of the DataFrame to work with
stance = fights.copy()

# Define a function to determine the winner's stance based on the corner
def determine_winner_stance(row):
    if row['Winner By Corner'] == 'Red':
        return row['R_Stance']
    elif row['Winner By Corner'] == 'Blue':
        return row['B_Stance']
    else:
        return 'Draw'

# Apply the function to create the 'Winner By Stance' column
stance['Winner By Stance'] = stance.apply(determine_winner_stance, axis=1)

# Perform the count using pandas
winner_stance_counts = stance['Winner By Stance'].value_counts()

# Create a DataFrame from the counts
df_winner_stance_counts = pd.DataFrame({'Winner By Stance': winner_stance_counts.index, 'Total Wins': winner_stance_counts.values})

# Sort the DataFrame based on 'Total Wins' in descending order for consistent coloring
df_winner_stance_counts.sort_values(by='Total Wins', ascending=False, inplace=True)

# Create the Plotly bar graph using px.bar and manually set the colors using Inferno color scale
fig = px.bar(df_winner_stance_counts,
             x='Winner By Stance',
             y='Total Wins',
             color='Total Wins', # Use 'Total Wins' for coloring
             color_continuous_scale='Bluered_r', # Use Inferno color scale
             category_orders={'Winner By Stance': df_winner_stance_counts['Winner By Stance']}, # Ensure consistent coloring
             labels={'Total Wins': 'Total Wins', 'Winner By Stance': 'Winner By Fighting Stance'}) # Set axis labels

# Set the axis labels and title
fig.update_layout(title='Count of Winners by Fighting Stance')

# Show the plot
fig.show()

```

```python
# SUB = fights[['R_SUB_ATT','B_SUB_ATT','Winner By Corner','Winner','Year']]

# # five_years_ago_year = 2021 - 5

# # TD = TD[TD['Year'] >= five_years_ago_year]

# SUB['SUB_diff'] = SUB.apply(lambda row: row['R_SUB_ATT'] - row['B_SIG_TD'] if row['Winner By Corner'] == 'Red' 
#                                  else (row['B_SIG_TD'] - row['R_SIG_TD'] if row['Winner By Corner'] == 'Blue' else 0), axis=1)
# import plotly.express as px
# fig = px.scatter(TD, x="TD_diff", y="Winner", color="TD_diff",
#                  labels={"TD_diff": "Take Down", "Winner": "Fighter"},
#                  title="Take Down Difference between Fighters by Winner")
# fig.show()

SUB = fights.copy()
def winner(row):
    if row['Winner By Corner'] == 'Red' and row['R_SUB_ATT'] > row['B_SUB_ATT']:
        return "Winner with more SUB"
    elif row['Winner By Corner'] == 'Blue' and row['B_SUB_ATT'] > row['R_SUB_ATT']:
        return "Winner with more SUB"
    elif row['Winner By Corner'] == 'Red' and row['R_SUB_ATT'] < row['B_SUB_ATT']:
        return "Winner with less SUB"
    elif row['Winner By Corner'] == 'Blue' and row['B_SUB_ATT'] < row['R_SUB_ATT']:
        return "Winner with less SUB"
    elif row['R_SUB_ATT'] == row['B_SUB_ATT']:
        return "Same attemp"
    elif row['B_SUB_ATT'] == 0 and row['R_SUB_ATT'] == 0:
        return "No SUB"
    elif row['Winner By Corner'] == 'Draw':
        return 'Draw'
    else:
        return 'No Streak'

SUB['submission'] = SUB.apply(winner, axis=1)
a = SUB.groupby('submission')['Winner'].count()
a
```

```python
streak = fights.copy()
def winner(row):
    if row['Winner By Corner'] == 'Red' and row['R_current_win_streak'] > row['B_current_win_streak']:
        return "Winner with more streak"
    elif row['Winner By Corner'] == 'Blue' and row['B_current_win_streak'] > row['R_current_win_streak']:
        return "Winner with more streak"
    elif row['Winner By Corner'] == 'Red' and row['R_current_win_streak'] < row['B_current_win_streak']:
        return "Winner with less streak"
    elif row['Winner By Corner'] == 'Blue' and row['B_current_win_streak'] < row['R_current_win_streak']:
        return "Winner with less streak"
    elif row['B_current_win_streak'] == row['R_current_win_streak']:
        return "Same streak"
    elif row['B_current_win_streak'] == 0 and row['R_current_win_streak'] == 0:
        return "No Streak"
    elif row['Winner By Corner'] == 'Draw':
        return 'Draw'
    else:
        return 'No Streak'

streak['winning_streak'] = streak.apply(winner, axis=1)
a = streak.groupby('winning_streak')['Winner'].count()
a
```
